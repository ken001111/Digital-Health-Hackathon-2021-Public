# Calculate Jaccard Coefficient between genetic alterations

- 유전자 변이 정보는 1을 가지는 데이터가 0을 가지는 데이터보다 희소하다.
- 따라서 희소 데이터의 유사도를 측정할 때 사용하는 Jaccard Coefficient를 이용하여 유전자 변이 간 상관 관계를 찾으려 한다.

```python
import pandas as pd
```

### Genetic_alterations.csv
- 유전자 변이 정보가 들어있는 데이터

```python
gene_mutations = pd.read_csv("Genetic_alterations.csv")
```

```python
gene_mutations.head()
```

- Jaccard Coefficient를 저장할 300 * 300 Dataframe 생성

```python
list_jaccard = pd.DataFrame([[0 for i in range(300)] for j in range(300)], columns=['G' + str(i+1) for i in range(300)], index=['G' + str(i+1) for i in range(300)], dtype='float')
```

```python
list_jaccard
```

- 두 유전자 변이 A, B에 대해서 (A와 B 모두 1인 환자 수) / (A와 B 둘 중 하나라도 1인 환자 수) = 자카드 계수 계산

```python
for A in range(1, 301):
  for B in range(A, 301):
    list_A = gene_mutations['G' + str(A)].values
    list_B = gene_mutations['G' + str(B)].values
    
    numerator = 0
    denominator = 0
    
    for k in range(len(list_A)):
      if list_A[k] == 1 and list_B[k] == 1:
        numerator += 1
        denominator += 1
      elif list_A[k] == 1 or list_B[k]:
        denominator += 1
    jaccard = numerator / denominator
    
    list_jaccard['G' + str(A)][B-1] = jaccard
```

- Save as csv file

```python
list_jaccard.to_csv("jaccard_coefficient.csv")
```
