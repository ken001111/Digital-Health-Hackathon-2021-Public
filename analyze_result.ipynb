# c-index 가 높은 순으로 정렬
- 결과 확인

```python
result.sort_values(by=['c index'], ascending=False)
```

# 기준 c-index 값보다 작은 행 삭제
- list_for_del : 조건에 맞지 않은 행의 인덱스를 저장해놓은 리스트
- standard : 모델이 적합한지 판단하는 c-index의 기준
- 모델이 기준에 적합하지 않으면, list_for_del에 저장한 뒤 drop 함수를 통해 전체 dataframe에서 삭제

```python
list_for_del=[]
```

```python
standard = 0.7
```

```python
for i in range(len(result)):
  if result['c index'].values[i] < standard:
      list_for_del.append(i)
```

```python
result_passed = result.drop(list_for_del)
```

- 기준을 만족하는 DataFrame 출력

```python
result_passed
```

# 치료효능률 평균 계산

```python
ratio_average = []
for i in range(300):
  ratio_average.append([result_passed['G' + str(i+1)].values.mean(), 'G' + str(i+1)])
ratio_average.sort()
```

```python
ratio_average[:10]
```

# 유전자 후보 10개 제시
- 유효한 모델 결과에 대해 평균 치료효능률이 작은 순으로 유전자 10개 출력

```python
candidate = [i[1] for i in ratio_average[:10]]
print('유전자 후보: ' + str(candidate))
```
