# PyTorch 역전파

https://www.youtube.com/watch?v=_4gEHlsvSZ8

```python
import torch
```

```python
# requires_grad = False가 기본값이지만, 
# True로 설정할 경우 해당 변수를 미분할 수 있게됨
x = torch.ones(2,2, requires_grad=True)
print(x)
```

```python
y = x + 1
print(y)
```

```python
# z = 2 X y^2
z = 2 * y ** 2
print(z)
```

- z = 2 y ^ 2 
- z = 2 (x + 1) ^ 2
- dz/dx = 4 (x + 1)
- 따라서 z.backward(gradient=torch.ones(2,2)) 하면 x.grad에 4 (x + 1) 값이 저장된다. 

```python
# z.backward(gradient=torch.ones(2,2))
# print(x.grad)
# tensor([[8., 8.],
#       [8., 8.]])
```

```python
z.backward(gradient=torch.ones(2,2))
```

```python
print(x.grad)
```

```python
# 유튜브에서는 gradient 인자로 넘겨주지 않고 평균을 구하는 방식으로 했음.
res = z.mean()
print(res)
```

```python
# d(res)/dx_i 는?
# res = (z_1 + z_2 + z_3 + z_4)/4
# d(res)/dx = x + 1 이 된다.
res.backward()
```

```python
print(x.grad)
```

# Autograd에서 미분

https://tutorials.pytorch.kr/beginner/blitz/autograd_tutorial.html

```python
import torch
a = torch.tensor([2.,3.], requires_grad = True)
b = torch.tensor([6.,4.], requires_grad = True)
print(a)
```

```python
Q = 3 * a ** 3 - b ** 2
print(Q)
```

```python
# Q : 오차(error), 종속변수
# a, b : 매개변수, 독립변수
# Q = 3a^3 - b^2
# Q를 a에 대해 편미분 = 9a^2
# Q를 b에 대해 편미분 = -2b
```

```python
# Q가 어떤 값이 아니라 [x, y] 형태의 벡터이므로
# Q.backward()를 할 때 인자를 넘겨줘야함. (gradient=???)
# 인자로는 Q와 같은 형태의 tensor를 넘겨주면 된다.
external_grad = torch.tensor([1., 1.,])
Q.backward(gradient=external_grad)
Q.backward()
```

```python
# a.grad에는 dQ/da 값 (= 9a^2) 이 저장되있는 격.
# 따라서 다음 두 결과는 같다.
print(a.grad)
print(9*a**2)
print(b.grad)
```
