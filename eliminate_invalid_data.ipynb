# 유효하지 않은 환자 데이터 검색
- list_for_del : 유효하지 않은 환자 번호(index)를 저장할 리스트
- for loop을 돌면서 임상변수의 조건 (0보다 크거나 같고 9보다 작거나 같다)을 만족하지 않는 환자 번호를 저장
- for loop을 돌면서 생존시간이 음수인 환자 번호를 저장

```python
list_for_del=[]
```

```python
for i in range(1,11):
  for j in range(1000):
    if dataset['Var'+str(i)].values[j]>=10:
      list_for_del.append(j)
```

```python
for i in range(1000):
  if dataset['time'].values[i]<0:
    list_for_del.append(i)
```

- 겹치는 번호가 없는 지 확인 후 정렬

```python
print(len(list_for_del))
print(len(list(set(list_for_del))))
```

```python
list_for_del.sort()
```

- 유효하지 않은 환자 데이터를 제거
- 유효하지 않은 환자 수 : 34명
- 유효한 환자 수 : 966명

```python
dataset_modified = dataset.drop(list_for_del)
```

```python
dataset_modified
```

- 변경된 DataFrame을 csv파일로 저장

```python
dataset_modified.to_csv('/content/all-in-one-modified.csv', index= False)
```
